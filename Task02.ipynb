# Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
sns.kdeplot(df['Sales'],color='red',label='Sales',shade=True,bw=25)
sns.kdeplot(df['Profit'],color='Blue',label='Profit',shade=True,bw=25)
plt.xlim([-100,1000])
plt.legend
```

# Analysis using pairplot of each column

```python
import seaborn as sns
sns.pairplot(df,hue='Category')
```

```python
import seaborn as sns
sns.heatmap(df.corr(),cmap='rocket_r',annot=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))
sns.countplot(df['Sub-Category'])
plt.title('Sub-Category',fontsize=20)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(18,5))
sns.countplot(df['Discount'])
plt.xticks(rotation=90)
plt.title('Discount',fontsize=20)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(18,5))
sns.countplot(df['Quantity'])
plt.title('Quantity',fontsize=20)
```

# Distribution of the data using plot

```python
fig , axs = plt.subplots(ncols=2,nrows=2,figsize=(10,10))
sns.displot(df['Sales'],color='blue',ax=axs[0][0])
sns.displot(df['Profit'],color='red',ax=axs[0][1])
sns.displot(df['Quantity'],color='green',ax=axs[1][0])
sns.displot(df['Discount'],color='orange',ax=axs[1][1])
axs[0][0].set_title('Sales Distrubtion')
axs[0][1].set_title('Profile Distrubtion')
axs[1][0].set_title('Quantity Distrubtion')
axs[1][1].set_title('Discount Distrubtion')
```

```python
df['Segment'].value_counts()
```

```python
df_segment=df.groupby(['Segment'])[['Sales','Discount','Profit']].mean()
df_segment
```

```python
import matplotlib.pyplot as plt
df_segment.plot.pie(subplots=True, autopct='%1.1f%%', figsize=(18,20), startangle=90, shadow=True, labels=df_segment.index)
plt.title('Segment wise analysis of Sale,Discount,Profit')
```

# Region wise Analysis

```python
df_region=df.groupby(['Region'])[['Sales','Profit','Discount']].mean()
df_region
```

```python
import matplotlib.pyplot as plt
df_region.plot.pie(subplots=True, autopct='%1.1f%%', figsize=(18,20), labels=df_region.index)
```
